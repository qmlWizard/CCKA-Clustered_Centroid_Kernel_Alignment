```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from utils.data_generator import DataGenerator
```

```python
data_generator = DataGenerator(
        dataset_name='Donuts',
        file_path= "data\donuts.npy",
    )

training_data, training_labels, testing_data, testing_labels = data_generator.generate_dataset()
fig = data_generator.plot_dataset(training_data, training_labels, testing_data, testing_labels)
fig.savefig("plots/donuts_dataset.png", dpi=300, bbox_inches="tight")

clf = SVC(kernel='rbf', C=1.0, gamma='scale')  # 'scale' = 1 / (n_features * X.var())
clf.fit(training_data, training_labels)

# Predictions
y_pred = clf.predict(testing_data)

# Evaluation
print("Accuracy:", accuracy_score(testing_labels, y_pred))
print("\nClassification Report:\n", classification_report(testing_labels, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(testing_labels, y_pred))
```
