```python
import json 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
print(plt.style.available)
```

***Checkerboard Data***

```python
with open('he_results/checkerboard_ccka_4_exp1.json') as file:
    ccka_data = json.load(file)

with open('he_results/checkerboard_random_8_exp1.json') as file:
    random_data = json.load(file)

ccka_runs = ccka_data["runs"]
random_runs = random_data["runs"]

ccka_init_acc = []
ccka_final_acc = []

random_init_acc = []
random_final_acc = []

for i in range(5):
    ccka_init_acc.append(ccka_runs[i]['init']['accuracy'])
    random_init_acc.append(random_runs[i]['init']['accuracy'])

    
    ccka_final_acc.append(ccka_runs[i]['final']['accuracy'])
    random_final_acc.append(random_runs[i]['final']['accuracy'])

ccka_init_avg = np.mean(np.array(ccka_init_acc))
ccka_final_avg = np.mean(np.array(ccka_final_acc))

random_init_avg = np.mean(np.array(random_init_acc))
random_final_avg = np.mean(np.array(random_final_acc))

algorithms = ['CCKA', 'Random']
x = np.arange(len(algorithms))
# Keeping the bar width smaller and reducing distance between bars by adjusting positions
plt.figure(figsize=(10, 6))
width = 0.25  # Keeping the original smaller bar width

# Use a clean Seaborn style
plt.style.use('seaborn-v0_8')

# Plotting the average initial and final accuracy with less distance between bars
bars1 = plt.bar(x - width/2, [ccka_init_avg, random_init_avg], width, label='Initial Accuracy', color='#ff7f0f', edgecolor='black', alpha=0.8)
bars2 = plt.bar(x + width/2, [ccka_final_avg, random_final_avg], width, label='Final Accuracy', color='#1f77b4', edgecolor='black', alpha=0.8)

# Adding values on top of each bar
for bar in bars1:
    plt.text(bar.get_x() + bar.get_width()/4, bar.get_height() + 0.01, f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=9, fontweight='bold')

for bar in bars2:
    plt.text(bar.get_x() + bar.get_width()/4, bar.get_height() + 0.01, f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=9, fontweight='bold')

# Adding labels, title, and legend
plt.xlabel('Algorithm', fontsize=12, fontweight='bold')
plt.ylabel('Accuracy', fontsize=12, fontweight='bold')
plt.xticks(x, algorithms, fontsize=11)
plt.title('Average Initial and Final Accuracy for CCKA and Random Algorithms (CheckerBoard dataset)', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='best')

# Make axis spines less visible for a clean look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add grid lines for y-axis to make it easier to compare heights
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Display the chart
plt.tight_layout()
plt.savefig('accuracy_checkerboard_he.png', dpi=300, bbox_inches='tight')  # Save with 300 DPI for high quality
plt.show()


plt.plot(algorithms, [ccka_runs[i]['final']['executions'], random_runs[i]['final']['executions']], marker='o', linestyle='-', color='#1f77b4', linewidth=2, markersize=8)

# Adding values on each point
for i, value in enumerate([ccka_runs[i]['final']['executions'], random_runs[i]['final']['executions']]):
    plt.text(i, value + 200, f'{int(value)}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Adding labels, title, and grid lines
plt.xlabel('Algorithm', fontsize=12, fontweight='bold')
plt.ylabel('Number of Executions', fontsize=12, fontweight='bold')
plt.xticks(fontsize=11)
plt.title('Average Number of Executions for CCKA and Random Algorithms', fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Make axis spines less visible for a clean look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Display the chart
plt.tight_layout()
plt.savefig('executions_checkerboard_he.png', dpi=300, bbox_inches='tight')  # Save with 300 DPI for high quality
plt.show()
```
